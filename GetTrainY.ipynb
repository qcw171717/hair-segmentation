```python
# http://vis-www.cs.umass.edu/lfw/part_labels/
import os
from PIL import Image
import numpy as np
```

```python
curr_dir = %pwd
images_directory = curr_dir  + '/parts_lfw_funneled_gt_images'
files = []
for root, dirs, f in os.walk(images_directory): 
    files.extend(f)

test_dir = images_directory + '/' + files[0]
im = Image.open(test_dir)
this_ar = np.array(im)
image_map = {}
for file in files:
    this_dir = images_directory + '/' + file
    im = Image.open(this_dir)
    this_array = np.array(im)
    image_map[file] = this_array
```

```python
import pickle
with open('TrainY.pkl', 'wb') as f:
    pickle.dump(image_map, f)
```
